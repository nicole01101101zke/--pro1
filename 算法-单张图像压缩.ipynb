```python
import numpy as np
import imageio

def func(x):
    average = np.mean(x,axis=0)
    m, n = np.shape(x)
    data_adjust = []
    avgs = np.tile(average, (m, 1))
    img_adjust = x - avgs
    eig_val,eig_vec = np.linalg.eig(np.cov(img_adjust))
    sigma = np.argsort(eig_val)
    sigma= sigma[::-1]
    eig_vec = eig_vec[:,sigma]
    eig_val = eig_val[sigma]
    
    sum = np.sum(eig_val)
    rk=np.linalg.matrix_rank(x)
    part = 0
    for i in range(rk):
        part = part + eig_val[i]
        if(part > 0.9*sum):
            break;
    eig_val_choose = i
    eig_vec = eig_vec[:,range(eig_val_choose)]
    
    #计算空间节省，压缩率
    score = np.dot(eig_vec.T,img_adjust)
    new = score.shape[0]*score.shape[1]
    old = m*n
    comp = 1.0*new/old
    print(comp)
    
    #计算重构误差，案例研究
    recon = np.dot(eig_vec,score)+np.mean(x,axis=0).T
    #print(recon)
    
    err=np.linalg.norm(recon-x)
    print(err)
    result = np.uint8(np.absolute(recon))
    return result
    
    

for i in range(0,100):
    if i <=9:
        img = imageio.imread(r"D:/大三上/数据科学与工程算法/project1/Images/beach/beach0"+str(i)+".tif")
    else:
        img = imageio.imread(r"D:/大三上/数据科学与工程算法/project1/Images/beach/beach"+str(i)+".tif")
    #img = imageio.imread(r"D:/大三上/数据科学与工程算法/project1/Images/beach/beach00.tif")
    img_np = np.array(img)
    #print(img_np)
    img_r = img_np[:,:,0]
    img_g = img_np[:,:,1]
    img_b = img_np[:,:,2]
    #print(img_r)
    #print(img_g)
    #print(img_b)

    img_r_recon,img_g_recon,img_b_recon = func(img_r),func(img_g),func(img_b)
    img_recon = np.dstack([img_r_recon,img_g_recon,img_b_recon])
    imageio.imwrite(r"D:/大三上/数据科学与工程算法/project1/result/"+str(i)+".tif",img_recon)
   
    #for i in range(0,100):
#    if i <=9:
#        img = imageio.imread(r"D:/大三上/数据科学与工程算法/project1/Images/agricultural/agricultural0"+str(i)+".tif")
#    else:
#        img = imageio.imread(r"D:/大三上/数据科学与工程算法/project1/Images/agricultural/agricultural"+str(i)+".tif")
#    #img = imageio.imread(r"D:/大三上/数据科学与工程算法/project1/Images/beach/beach00.tif")
#    img_np = np.array(img)
#    #print(img_np)
#    img_r = img_np[:,:,0]
#    img_g = img_np[:,:,1]
#    img_b = img_np[:,:,2]
#    #print(img_r)
#    #print(img_g)
#    #print(img_b)
#
#    img_r_recon,img_g_recon,img_b_recon = func(img_r),func(img_g),func(img_b)
#    img_recon = np.dstack([img_r_recon,img_g_recon,img_b_recon])
#    imageio.imwrite(r"D:/大三上/数据科学与工程算法/project1/result/"+str(i)+"agricultural.tif",img_recon)
   
#for i in range(0,100):
#    if i <=9:
#        img = imageio.imread(r"D:/大三上/数据科学与工程算法/project1/Images/airplane/airplane0"+str(i)+".tif")
#    else:
#        img = imageio.imread(r"D:/大三上/数据科学与工程算法/project1/Images/airplane/airplane"+str(i)+".tif")
#    #img = imageio.imread(r"D:/大三上/数据科学与工程算法/project1/Images/beach/beach00.tif")
#    img_np = np.array(img)
#    #print(img_np)
#    img_r = img_np[:,:,0]
#    img_g = img_np[:,:,1]
#    img_b = img_np[:,:,2]
#    #print(img_r)
#    #print(img_g)
#    #print(img_b)
#
#    img_r_recon,img_g_recon,img_b_recon = func(img_r),func(img_g),func(img_b)
#    img_recon = np.dstack([img_r_recon,img_g_recon,img_b_recon])
#    imageio.imwrite(r"D:/大三上/数据科学与工程算法/project1/result/"+str(i)+"airplane.tif",img_recon)
```
